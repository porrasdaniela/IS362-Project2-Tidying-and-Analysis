Dataset 3: Aaron Judge Data Cleaning and Analysis

In this notebook, we will clean and analyze a dataset featuring key baseball statistics for Aaron Judge, focusing on both his 2024 season and overall career totals. The dataset includes metrics such as at-bats (AB), home runs (HR), batting average (BA), on-base percentage (OBP), slugging percentage (SLG), and more. This dataset was shared by my classmate Nicholas Velasquez. In the discussion, he suggested analyzing performance trends across Aaron Judge’s career, including the evolution of batting average, home run counts, and key metrics like OBP and SLG. We will also explore potential patterns or correlations between games played and home run totals to better understand Judge’s development and future performance potential.

1. Import Libraries & Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the CSV file
df = pd.read_csv("aaron_judge_data.csv")

# Display the first few rows
df.head()
```

Here we loaded Aaron Judge's stats CSV into pandas and inspected the data. 

2. Tidy & Transform Data

```python
# Convert Year to string type since 'Career' is text
df['Year'] = df['Year'].astype(str)

# Ensure all other columns except 'Year' are numeric
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Check datatypes
df.dtypes
```

We converted the 'Year' column to a string since it includes the text "Career." We then made sure all the performance-related columns (WAR, AB, HR, etc.) are treated as numeric data types using pd.to_numeric(). This will allow us to perform mathematical and statistical analyses on these columns without issues.

3. Basic Data Cleaning

```python
# Check for missing data
df.isnull().sum()
```

Here we performed a missing data check and confirmed that there are no missing values in the dataset. This means we don't need to perform imputation or further data cleaning for missing values.

4. Perform the requested analyses given by my classmate

```python
# Analysis 1 - Trends in Batting Average, HR, OBP, and SLG

# Plot batting average, OBP, SLG over Year
plt.plot(df['Year'], df['BA'], marker='o', label='Batting Average')
plt.plot(df['Year'], df['OBP'], marker='o', label='OBP')
plt.plot(df['Year'], df['SLG'], marker='o', label='SLG')
plt.title("Judge's Batting Avg, OBP, SLG Trends")
plt.xlabel("Year")
plt.ylabel("Metrics")
plt.legend()
plt.grid()
plt.show()
```

This line plot shows the trends in Aaron Judge's Batting Average (BA), On-Base Percentage (OBP), and Slugging Percentage (SLG) between his 2024 season and his career average.

The batting average dropped slightly when comparing his 2024 season to his career average.
Similarly, both OBP and SLG saw slight decreases in his career totals compared to the 2024 season.
This could suggest that 2024 was one of Judge’s stronger offensive seasons relative to his career so far, especially in terms of power metrics like slugging.

```python
# Analysis 2: Home Runs vs. At Bats

plt.scatter(df['AB'], df['HR'])
plt.title("Home Runs vs. At Bats")
plt.xlabel("At Bats (AB)")
plt.ylabel("Home Runs (HR)")
plt.grid()
plt.show()
```

This scatterplot compares Aaron Judge's Home Runs (HR) against his At Bats (AB) across his 2024 season and his career totals.

The data suggests a strong positive relationship between at-bats and home run totals, as expected. More at-bats typically give players more chances to hit home runs.
His career point (3564 AB & 315 HR) is significantly higher than his 2024 season data point (559 AB & 58 HR), showing the accumulation over time.
However, based on the ratio, Judge's 2024 season appears to have been highly productive in terms of home runs relative to his number of at-bats.

```python
# Analysis 3: RBI vs. Runs (R)
plt.scatter(df['R'], df['RBI'])
plt.title("Runs vs. RBIs")
plt.xlabel("Runs (R)")
plt.ylabel("Runs Batted In (RBI)")
plt.grid()
plt.show()
```

This scatterplot compares Aaron Judge's Runs (R) to his Runs Batted In (RBI) for both the 2024 season and his career totals.

The plot indicates a strong positive relationship between runs and RBIs. Generally, players who score more runs tend to also drive in more RBIs, showing their overall offensive productivity.
Judge's career numbers (736 R and 716 RBI) and 2024 season (122 R and 144 RBI) both reinforce this pattern, with a slightly higher RBI count relative to runs scored in 2024, suggesting he played a larger role in driving in teammates this season.
